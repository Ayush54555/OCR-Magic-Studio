# Import Dependencies

```python
import easyocr
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Read images or videos

```python
img_path = 'samp_1.jpg'
img_path2 = 'samp_2.png'
```

```python
reader = easyocr.Reader(['en','hi'])
result = reader.readtext(img_path)
result2 = reader.readtext(img_path2)
result
```

# Draw Results

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(img_path)
img = cv2.rectangle(img, top_left, bottom_right, (0,255,0), 5)
img = cv2.putText(img, text, top_left, font, 0.8, (255,0,255), 2, cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

# Handling Multiple Lines

```python
img = cv2.imread(img_path2)
for detection in result2:
    top_left = tuple([int(val) for val in detection[0][0]])
    bottom_right = tuple([int(val) for val in detection[0][2]])
    text = detection[1]
    font = cv2.FONT_HERSHEY_SIMPLEX
    img = cv2.rectangle(img, top_left, bottom_right, (0,255,0), 2)
    img = cv2.putText(img, text, top_left, font, 0.4, (255,0,0), 1, cv2.LINE_AA)
    
plt.imshow(img)
plt.show()
```
